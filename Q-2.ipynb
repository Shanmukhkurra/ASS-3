```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("/content/diabetes.csv")
population_bmi = data['BMI']
population_98th_percentile = population_bmi.quantile(0.98)
print("Population 98th Percentile BMI:", population_98th_percentile)
sample_size = 100
sample = population_bmi.sample(sample_size, random_state=42)
sample_98th_percentile = sample.quantile(0.98)
print("Sample 98th Percentile BMI:", sample_98th_percentile)
plt.figure(figsize=(10, 7))
plt.hist(population_bmi, bins=30, alpha=0.7, label='Population')
plt.axvline(population_98th_percentile, color='green', linestyle='dotted', linewidth=2.5, label='Population 98th Percentile')
plt.hist(sample, bins=30, alpha=0.7, label='Sample')
plt.axvline(sample_98th_percentile, color='blue', linestyle='dotted', linewidth=2.5, label='Sample 98th Percentile')
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.title('BMI Distribution and 98th Percentile Comparison')
plt.legend()
plt.show()

```
